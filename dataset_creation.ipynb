```python
import re

data = []

with open("restaurants.sql", "r") as file:
    for line in file:
        match = re.search(r"\((.*?)\)", line)
        if match:
            values = match.group(1).split(",")
            values = [v.strip().replace("'", "") for v in values]
            data.append(values)

restaurants = pd.DataFrame(data, columns=[
    "restaurant_id", "restaurant_name", "cuisine", "rating"
])

restaurants["restaurant_id"] = restaurants["restaurant_id"].astype(int)

print(restaurants.head())
```

```python
restaurants["rating"] = restaurants["rating"].astype(float)
```

```python
restaurants.info()
```

# Step 2 : Merging the data files in order to create our dataset ie final_food_delivery_dataset.csv

```python
df = pd.merge(
    orders,
    users,
    on="user_id",
    how="left"   # Keep all orders
)
```

```python
df = pd.merge(
    df,
    restaurants,
    on="restaurant_id",
    how="left"
)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["order_date"] = pd.to_datetime(df["order_date"] , format='%d-%m-%Y')
```

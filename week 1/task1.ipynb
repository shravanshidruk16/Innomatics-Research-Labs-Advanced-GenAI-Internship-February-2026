### 1. User Login Check

```python
username = input('Enter the username: ')
password = input('Enter the password: ')
if username=='admin' and password=='1234':
    print('Login Successful!')
else:
    print('Invalid Credentials')
```

### 2. Pass / Fail Analyzer

```python
marks = [45, 78, 90, 33, 60]
pass_students = 0
fail_students = 0
for i in range(0,len(marks)):
    if marks[i]>=50:
        pass_students += 1
    else:
        fail_students += 1
print(f'1.The total number of PASSED students = {pass_students}\n2.The total number of FAILED students = {fail_students}')
```

### 3. Simple Data Cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]
for i in range(len(names)):
    names[i] = names[i].strip().lower()
print(names)
```

### 4. Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for msg in messages:
    msg_len = len(msg)
    print(f"Message : {msg} ---> Length : {len(msg)}")
    if msg_len>10:
        print("ABOVE MESSAGE IS LONGER THAN 10 CHARACTERS !")
```

### 5. Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0
for log in logs:
    if log=='ERROR':
        error_count += 1
print("The total number of ERROR logs = ", error_count)
```

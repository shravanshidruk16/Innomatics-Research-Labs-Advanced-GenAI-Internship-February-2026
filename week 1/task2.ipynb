### 1. Problem 1: Unique Words in a Sentence

```python
user_input = input("Enter a sentence: ")
unique_words = set(user_input.split(' '))
print(f"The count of unique words is : {len(unique_words)}")
print(f"The unique words are : {unique_words}")
```

### 2. Problem 2: Highest Salary from Employee Data

```python
employees = {} # employees dict
n = int(input('Enter the total no. of employees you want to add : ')) # not hardcoding the employees data
for i in range(n):
    name = input('Enter the name : ')
    salary = int(input('Enter the salary: '))
    employees[name] = salary
maximum_sal = 0 # to find maximum salary
emp_with_max_sal = '' # to find person with maximum salary
for name in employees:
    if employees[name] > maximum_sal:
        maximum_sal = employees[name]
        emp_with_max_sal = name
print(f"Highest salary : {emp_with_max_sal} = {maximum_sal}")
```

```python
# Here we created a dictionary then implemented this python one-liner to find the employee with maximum salary
print(f"Highest salary : {str(max(employees))} = {employees[max(employees)]}")
```

### 3. Problem 3: Find Maximum and Minimum Values

```python
n = int(input('Enter the TOTAL count of numbers you want to add in list : '))
numbers = []
for i in range(n):
    s = int(input('Enter the number : '))
    numbers.append(s)
max_no = numbers[0]
min_no = numbers[0]
for num in numbers:
    if num>max_no:
        max_no = num
    if num<min_no:
        min_no = num
print(f"List : {numbers}\nMaximum number : {max_no}\nMinimum number : {min_no}")
```

### 4. Problem 4: Count Products Above a Price Threshold

```python
tot_products = []
n = int(input("Enter the TOTAL count of products you want to add in cart : "))
for i in range(n):
    s = int(input(f'Enter the price of the product {i+1}: '))
    tot_products.append(s)
costly_products = 0
for product in tot_products:
    if product>1000:
        costly_products += 1
print(f"Products above 1000 : {costly_products}")
```

### 5. Problem 5: Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]
present = 0
for at in attendance:
    if at=="P":
        present += 1
precentage = round(float((present*100)/len(attendance)),2)
print(f"Attendance Percentage : {precentage} %")
```

### 7. Problem 7: Count Character Frequency

```python
user_inp = input("Enter the string : ")
freq = {}
for ch in user_inp:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1

print(freq)
```

### 9. Problem 9: Check if a Key Exists in a Dictionary

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
employee_check = input("Enter the employee name : ")
if employee_check in employees:
    print(f"Employee Exists !!!\nEmployee : {employee_check} is PRESENT with salary : {employees[employee_check]}")
else:
    print(f"Employee DOES NOT EXISTS !!!\nEmployee : {employee_check} is NOT PRESENT")
```

### 10. Problem 10: Calculate Average Marks

```python
marks = [85, 92, 78, 90, 88, 95, 80, 91, 89, 94]
sum_of_marks = 0
for mark in marks:
    sum_of_marks += mark
average = float(sum_of_marks / len(marks))
print(f"Average Marks = {average}")
print(f"One liner to calculate average : {sum(marks)/len(marks)}")
```

### 1. Problem Statement 1: Employee Performance Bonus Eligibility

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# find highest score
max_score = max(employees.values())

# find employees with max score
top_performers = [emp for emp, score in employees.items() if score == max_score]

# print in required format
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {max_score})")
```

### 2. Problem Statement 2: Search Query Keyword Analysis

```python
import string

user_input = input("Enter the string input: ")

# Convert to lowercase
user_input = user_input.lower()

# Remove punctuation
user_input = user_input.translate(str.maketrans('', '', string.punctuation))

# Split into words
words = user_input.split()

# Count frequency
freq = {}

for word in words:
    freq[word] = freq.get(word, 0) + 1

# Display only words searched more than once
result = {word : count for word , count in freq.items() if count > 1}

print(result)

```

### 3. Problem Statement 3: Sensor Data Validation

### 4. Problem Statement 4: Email Domain Usage Analysis

```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
gmail_count = 0
yahoo_count = 0 
for email in emails: 
    if email.endswith("@gmail.com"):
        gmail_count += 1
    else:
        yahoo_count += 1
tot_emails = len(emails)

print(f'Gmail Count (@gmail.com) : {round(gmail_count*100/tot_emails)}%\nYahoo Count (@yahoo.com) : {round(yahoo_count*100/tot_emails)}%')
```

### 6. Problem Statement 6: Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
freq = {}
for user in user_ids:
    if user in freq:
        freq[user] += 1
    else:
        freq[user] = 1
result = {user:count for user,count in freq.items() if freq[user]>1}
for user, count in result.items():
    print(f"{user} → {count} times")
```

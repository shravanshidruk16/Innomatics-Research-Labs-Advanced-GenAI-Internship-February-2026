### 1. Problem 1 : Mobile Recharge Validation System

```python
def recharge():
    amount = int(input('Enter the amount : '))
    while True:
        if amount<50:
            print('Error : Invalid Recharge Plan Entered ! ')
            amount=int(input('Please enter the amount again : '))
        else:
            if amount in [199,299,399,599]:
                print('Success : Valid Recharge Plan !')
                break
            else:
                amount=int(input('Valid Recharge Plans are [199,299,399,599] Enter an amount from them : '))
recharge()
```

### 2. Problem 2 : Inventory Reorder Alert System

```python
# Creating a dictionary with key as product name and value as quantity remanining in shop
stock_data = {
    "pen":20,
    "book":10,
    "pencil":30,
    "eraser":5,
    "whiteboard":10
}
for product,quantity in stock_data.items():
    if quantity < 15 :
        print(f'ALERT : {product.capitalize()} OUT OF STOCK , REORDER REQUIRED ! ')
    else : 
        print(f'OK : {product.capitalize()} stock is present')
        
```

### 3. Problem 3: Student Result Processing System

```python
def result_eval_sys(marks):
    avg_marks = sum(marks)/len(marks)
    if avg_marks>=50:
        return "Student PASSED"
    else:
        return "Student FAILED"

stud1_marks = [40,30,25,60,12]
stud2_marks = [50,67,72,39,80]
stud3_marks = [33,77,40,50,30]
print('Student 1 status : ',result_eval_sys(stud1_marks))
print('Student 2 status : ',result_eval_sys(stud2_marks))
print('Student 3 status : ',result_eval_sys(stud3_marks))
```

### 4. Problem 4: Cab Fare Estimator with Retry Option

```python
from datetime import datetime

peak_hours = [10,11,12,18,19,20,21] # selected peak hrs in 24 hrs format 

while True:
    base_fare = 50
    current_hr = datetime.now().hour
    
    km = int(input("Enter the distance in kilometers: "))
    ride_fare = 12 * km
    total_fare = base_fare + ride_fare
    
    if current_hr in peak_hours:
        total_fare += 0.25 * total_fare   # 25% extra on total fare
    
    print(f"Your total cab fare is: ₹{round(total_fare,2)}")
    
    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        print("Thank you for using my Cab Fare Estimator")
        break
```

### 5. Problem 5: Employee Attendance Eligibility Checker

```python
def hr_system(attendance_register):
    total_days = len(attendance_register)
    total_present = 0
    for attendance in attendance_register:
        if attendance=='P':
            total_present += 1
    tot_percentage = (total_present*100)/total_days
    if tot_percentage>=75:
        return f'Eligible : Percentage {tot_percentage} %'
    else : 
        return f'Not Eligible : Percentage {tot_percentage} %'
emp1 = ['P','P','P','A','A']
emp2 = ['P','A','P','A','A']
emp3 = ['P','A','P','P','A']
print(f'Employee 1 Attendance status -> {hr_system(emp1)}')
print(f'Employee 2 Attendance status -> {hr_system(emp2)}')
print(f'Employee 3 Attendance status -> {hr_system(emp3)}')
```

### 6. Problem 6 : Password Strength Checker

```python
def check_password_strength(password: str) -> str:
    # Rule 1: Minimum length
    if len(password) < 8:
        return "Weak: Password must be at least 8 characters long."
    
    has_digit = False
    has_special = False
    special_chars = "@#$"
    
    # Single loop over password
    for ch in password:
        if ch.isdigit():
            has_digit = True
        elif ch in special_chars:
            has_special = True
        
        # Early exit (optimization)
        if has_digit and has_special:
            return "Success : Your Password is STRONG !"
    
    # Final validation
    if not has_digit:
        return "Weak :  Password must contain at least one digit."
    if not has_special:
        return "Weak : Password must contain at least one special character (@#$)."
    
    return "Success : Your Password is STRONG !"


# Retry system
while True:
    pwd = input("Enter your password: ")
    result = check_password_strength(pwd)
    print(result)
    
    retry = input("Check another password? (yes/no): ").lower()
    if retry != "yes":
        print("Exiting Password Checker , thankyou")
        break
```

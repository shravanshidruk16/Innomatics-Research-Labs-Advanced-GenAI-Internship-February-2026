### 3. Problem 3: Agriculture – Rainfall Adequacy Checker

```python
past_users = {}

def rainfall_check(region, readings):
    past_users[region] = readings
    
    print("\n----- Rainfall Status -----")
    for region, readings in past_users.items():
        avg_readings = sum(readings) / len(readings)
        
        if avg_readings >= 75:
            print(f"STATUS OF {region} REGION : ADEQUATE RAINFALL ✅")
        else:
            print(f"STATUS OF {region} REGION : INADEQUATE RAINFALL ❌")
    print("----------------------------\n")


# Continuous Input Loop
while True:
    region = input("Enter Region Name (or type 'quit' to exit): ")
    
    if region.lower() == "quit":
        print("Program Ended.")
        break
    
    readings = []
    count = 1
    
    while count <= 5:
        value = float(input(f"Enter rainfall reading DAY {count} (in mm): "))
        readings.append(value)
        count += 1
    
    rainfall_check(region, readings)
```

### 4. Problem 4: Social Media – Duplicate Account Detection

```python
account_db = set()

def duplicate_check():
    n = int(input("Enter the total number of accounts you want to add : "))
    
    seen = set()
    duplicates = set()
    
    for i in range(n):
        user_id = input(f"Enter the User ID of #{i+1} account : ")
        
        if user_id in seen:
            duplicates.add(user_id)
        else:
            seen.add(user_id)

    if duplicates:
        for user in duplicates:
            print(f"Duplicate Account Found! : UserID = {user}")
    else:
        print("No duplicates found.")

duplicate_check()  
```

### 7. Problem 7: System Monitoring – Application Health Checker

```python
def app_monitor(performance):
    error_count = 0
    for errors in performance :
        if errors == -1 :
            error_count += 1
    if error_count==0:
        print(f"Healthy : Application is running smoothly , Errors = {error_count}")
    elif error_count <= 5:
        print(f"Minor issues : Application is running with minor issues , Errors = {error_count}")
    else:
        print(f"Critical issues : Application may stop due to critical issues , Errors = {error_count}")
performance = [random.choice([-1, 1]) for _ in range(10)]
app_monitor(performance)
```

### 9. Problem 9: E-Learning – Student Attendance Eligibility System

```python
import random
from typing import List

def attendance_system(student: List[str], tot_sessions: int) -> float:
    present_count = 0

    for status in student:
        if status == "P" :
            present_count += 1

    percentage = (present_count / tot_sessions) * 100

    if percentage >= 75:
        print("Eligible")
    else:
        print("NOT ELIGIBLE")

    return percentage


# Example usage
student1 = [random.choice(['P', 'A']) for _ in range(10)]
print("Attendance Record:", student1)
attendance_system(student1, len(student1))
```

### 10. Problem 10: Smart Electricity Bill Analyzer


```python
def generate_electricity_bill(total_units):
    final_cost = 0
    for i in range(1,total_units+1):
        if i<=100:
            final_cost += 3
        elif i>100 and i<=200:
            final_cost += 5
        else:
            final_cost += 7
    if final_cost < 500:
        print(f"Low Usage , Cost : ₹{final_cost}")
    elif final_cost>=500 and final_cost<1500:
        print(f'Moderate usage , Cost : ₹{final_cost}')
    else:
        print(f'HIGH ELECTRICITY USAGE!!! , Cost : ₹{final_cost}')
n = int(input(f'Enter count of houses you want to check : '))
while n>0:
    generate_electricity_bill(random.randint(100,500))
    n -= 1
```

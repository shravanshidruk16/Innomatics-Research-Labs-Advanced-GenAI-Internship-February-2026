### 1. Problem 1 : Smart Light Controller

```python
class LightAutomation:
    def __init__(self, area, action):
        self.area = area
        self.action = action

    def change_action(self):
        if self.action == "ON":
            self.action = "OFF"
        else:
            self.action = "ON"

    def display_status(self):
        print(f"{self.area} light is {self.action}!")

while True:
    print("Welcome to Light Automation System")
    print("1. Add a new light with its status ")
    print("2. Change a current light status")
    print("3. Quit")
    choice = int(input('Enter your choice : '))
    match choice:
        case 1:
            area = input("Enter the area in which light is located : ")
            status = input("Enter whether you want to turn it (ON/OFF) : ")
            user = LightAutomation(area,status)
            print("*"*25)
            user.display_status()
            print("*"*25)
        case 2:
            user.change_action()
            print("*"*25)
            user.display_status()
            print("*"*25)
        case 3:
            print("Thankyou for using my Light Automation System")
            break
```

### 2. Problem 2 : Employee ID Card System

```python
class Employee:
    def __init__(self,name,id,department):
        self.name = name
        self.id = id
        self.department = department
    def display_details(self):
        print("*"*10)
        print("Employee Details Management System")
        print(f"Name : {self.name}\nID : {self.id}\nDepartment : {self.department}")
        print("*"*10)

while True:
    print("1. Add Employee details")
    print("2. Find Employee")
    print("3. Quit")
    choice = int(input("Enter your choice number : "))
    match choice:
        case 1 : 
            emp_name = input("Enter the Employee Name : ")
            emp_id = input("Enter Employee's Company ID : ")
            emp_dept = input("Enter the department of the Employee : ")
            emp = Employee(emp_name,emp_id,emp_dept)
            print("Employee Added Successfully ! ")
        case 2 :
            emp.display_details()
        case 3  :
            print("Thankyou for using my Employee Details System!!!")
            break
```

### 3. Problem 3 : Mobile Contact Record

```python
class ContactRecords:
    def __init__(self,name,number):
        self.name = name
        self.number = number
    def display_contact(self):
        print(f"Name : {self.name}\nContact_no : {self.number}")

while True:
    print("1. Add a new contact")
    print("2. View details")
    print("3. Quit")
    choice = int(input("Enter Your Choice : "))
    match choice : 
        case 1 :
            name = input("Enter the name : ")
            number = input("Enter the number : ")
            new_contact = ContactRecords(name,number)
            print("New contact saved successfully!")
        case 2:
            print("*"*10)
            new_contact.display_contact()
            print("*"*10)
        case 3:
            print("Thankyou for using my Contact Record System")
            break
```

### 5. Problem 5 : Movie Rating Display System

```python
class MovieStatus:
    def __init__(self,name,ratings):
        self.name = name
        self.ratings = ratings
    def display_movie_status(self):
        print("*"*30)
        print("\tMOVIE DETAILS")
        print(f"Name : {self.name}\nRatings : {self.ratings}/5")
        print("*"*30)
while True:
    print("1. Add Movie Details")
    print("2. Show Movie Details")
    print("3. Quit")
    choice = int(input("Enter the choice : "))
    match choice : 
        case 1:
            name = input("Enter the movie name : ")
            ratings = input("Enter the movie ratings out of 5 stars : ")
            movie = MovieStatus(name,ratings)
            print("Movie Details Added Successfully!!!")
        case 2:
            movie.display_movie_status()
        case 3:
            print("Thankyou for using my Movie Rating Display System")
            break
```
